```python
columns = list(df.columns)
```

```python
columns.remove('What is the usual interval between your periods (Avg. duration of cycle)?')
```

```python
columns.remove('Age')
```

```python
columns
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
df = enc.fit
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('What is the usual interval between your periods (Avg. duration of cycle)?',axis=1)
y = df['What is the usual interval between your periods (Avg. duration of cycle)?']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(x_train,y_train)
```

```python
pred = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 1000)
```

```python
rfc.fit(x_train, y_train)
```

```python
pred = rfc.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=8)
```

```python
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
pred = nb.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
pred = svc.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```
